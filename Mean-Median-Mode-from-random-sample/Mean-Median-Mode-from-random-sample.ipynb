```
import pandas as pd
import numpy as np
%matplotlib inline
from scipy import stats
import matplotlib.pyplot as plt
```

```
result = pd.read_csv('https://gist.githubusercontent.com/shaswata56/7dbb4eaaa6851e1bad7a6eb9a20b1f61/raw/a9c0bb23d5dfdbac3fc59bf1a72f2fc067b70dc4/result.csv').values
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 20
fig_size[1] = 14
plt.rcParams["figure.figsize"] = fig_size
```

```
roll = result[0:result.shape[0], 0:1].reshape(result.shape[0])
marks = result[0:result.shape[0], 1:result.shape[1]].reshape(result.shape[0])
```

```
y_pos = np.arange(len(roll))

mean = np.mean(marks)
median = np.median(marks)
mode = stats.mode(marks)[0][0]
print('Mean: '+str(mean))
print('Median: '+str(median))
print('Mode: '+str(mode))

plt.bar(y_pos, marks, color='orange', label='Marks')
plt.xticks(y_pos, roll)
plt.axhline(mean, color='green', label='Mean')
plt.axhline(median, color='yellow', label='Median')
plt.axhline(mode, color='cyan', label='Mode')
plt.xlabel("Roll Number")
plt.ylabel("Marks")
plt.legend()
plt.show()
```
